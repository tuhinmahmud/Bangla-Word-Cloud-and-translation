```python
translator = Translator() # Create object of Translator.
word_list=[]
```

```python
bangla_word='ধন্যবাদ'
word=bangla_word
word_list.append(word)
```

```python
word_list=[]
word_list.append(bangla_word)
print("Number of languages:",len(googletrans.LANGUAGES))
for d in googletrans.LANGUAGES:
    #print(d)
    translated = translator.translate(bangla_word,dest=d)
    word_list.append(translated.text)
    #print(word_list)
print(word_list)
```

```python
#convert it to dictionary with values and its occurences
from collections import Counter
word_cloud_dict=Counter(word_list)
#print(word_cloud_dict)
myfontpath='SolaimanLipi-8-Jan-2011.ttf'
wordcloud = WordCloud(font_path=myfontpath,width = 1000, height = 500).generate_from_frequencies(word_cloud_dict)

plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
```
